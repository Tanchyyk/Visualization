```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("statistics.csv")
df["power"] = df["array_length"].apply(lambda x: int(np.log2(x)))
df.head()
```

```python
def plot_experiment_result(df):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    ax1.set_yscale('log')
    ax2.set_yscale('log')
    algorithm_names = list(df["algorithm_name"].unique())
    
    for algorithm_name in algorithm_names:
        cur_df = df[df["algorithm_name"] == algorithm_name]
        ax1.plot(cur_df["power"], cur_df["time"], label=algorithm_name, marker='o')
        ax2.plot(cur_df["power"], cur_df["compare_operations"], label=algorithm_name, marker='o')
    
    plt.legend()
    plt.show()
```

```python
plot_experiment_result(df[df["experiment"] == "experiment0"])
```

```python
plot_experiment_result(df[df["experiment"] == "experiment1"])
```

```python
plot_experiment_result(df[df["experiment"] == "experiment2"])
```
